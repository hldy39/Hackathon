```python
import yfinance as yf
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Hubert/OneDrive/Desktop/CompanyList.csv')
```

```python
print(df['Symbol'])
increased_symbols = [] 
# go thru ea of these stocks and cal the current volume that sold for ea of these stocks compared to previous period.
# everytime we find smtg that has an increased volume, add it to the list.
```

```python
# use a loop function to go thru each stock individually.
```

```python
for stock in df['Symbol']:
    stock = stock.upper() # ensure that tickers are in uppercase.
    if '^' in stock: # e.g. composite or index
        pass
    else: # create logic to check volitality.
        try:
            stock_info = yf.Ticker(stock) # create an obj which can be used to get info back from.
            hist = stock_info.history(period="5d") # history volume of 2nd last to the last 5 days.
            previous_averaged_volume = hist['Volume'].iloc[1:4:1].mean() # extracting the volume of the past 5 days.
            todays_volume = hist['Volume'][-1] # the last entry from the volume series (which is tdy's volume).
            if todays_volume > previous_averaged_volume * 2: 
                # python will notify when tdy's volume is double of previous avg volume.
                # it is possible to add other conditions using the 'and' statement.
                increased_symbols.append(stock)
        except:
            pass    
```

```python
print(increased_symbols)
```
